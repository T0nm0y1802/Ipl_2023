```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://www.timesofsports.com/cricket/ipl/squad-2023/'
r=requests.get(url)
print(r)
```

```python
data1=pd.read_csv('IPL2023_Match_Scoreboard.csv',encoding='unicode_escape')
data2=pd.read_csv('IPL2023_Matches.csv',encoding='unicode_escape')


data1.head()
```

```python
IPL2023_Matches_final=pd.merge(data1,data2,on='match_no')
```

```python
IPL2023_Matches_final.to_csv('IPL2023_Matches_final.csv')
```

```python
soup=BeautifulSoup(r.text,"lxml")
```

```python
print(soup)
```

```python
table=soup.find('table',dir='ltr')
```

```python
title=table.find_all('strong')
```

```python
header=[]
for i in title:
    name=i.text
    header.append(name)
    
header    
```

```python
df=pd.DataFrame(columns=header)
```

```python
row=soup.find_all('tr')
```

```python
for i in row:
    data=i.find_all("td")
    row=[tr.text for tr in data]
    l=len(df)
    df.loc[l]=row
    
print(df)    
```

```python
df.to_csv('player_info.csv')
```
